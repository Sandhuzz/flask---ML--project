##data preprocessing

```python
dt_cols = ["date_of_journey","dep_time","arrival_time"]
num_cols = ["duration","total_stops"]
cat_cols = [col for col in X_train.columns if (col not in dt_cols)    and (col not in num_cols)]            
```

```python
cat_cols
```

```python
num_transformer = Pipeline(steps =[
    ("imputer",SimpleImputer(strategy ="median")),
    ("scaler",StandardScaler())
])
cat_transformer = Pipeline(steps =[
    ("imputer",SimpleImputer(strategy ="most_frequent")),
    ("encoder",OneHotEncoder(sparse_output=False,handle_unknown ="ignore"))
])
doj_transfromer = Pipeline(steps =[
    ("imputer",SimpleImputer(strategy ="most_frequent")),
    ("extracter", DatetimeFeatures(features_to_extract = ["month","week","day_of_week","day_of_month"],format = "mixed")),
     ("scaler",StandardScaler())
])

time_transfromer = Pipeline(steps =[
    ("imputer",SimpleImputer(strategy ="most_frequent")),
    ("extracter", DatetimeFeatures(features_to_extract = ["hour","minute"],format = "mixed")),
     ("scaler",StandardScaler())
])
    
    
```

```python
preprocessor = ColumnTransformer(transformers = [
    ("num",num_transformer,num_cols),
    ("cat",cat_transformer,cat_cols),
    ("doj",doj_transfromer,["date_of_journey"]),
    ("time",time_transfromer,["dep_time","arrival_time"])
])
```

```python
preprocessor.fit_transform(X_train)
```

##model selection

```python
algorithms = {
    "Linesr Regression" : LinearRegression(),
    "Support vector machine " : SVR(),
    "Random Forest" :RandomForestRegressor(n_estimators = 10),
    "xg boost" : XGBRegressor(n_eztimators = 10)
}
```

```python
data = pd.concat([train_df,val_df],axis= 0)
X_data,y_data = split_data(data)
print(X_data.shape,y_data.shape )
```

```python
def plot_curve(sizes, mean_scores, std_scores, label, ax):
    ax.plot(sizes, mean_scores, label=label)
    ax.fill_between(
        sizes,
        mean_scores - std_scores,
        mean_scores + std_scores,
        alpha=0.3
    )
        
    
    
    
     
  

    
```

```python
def plot_learning_curve(name, algorithm, figsize=(12, 4)):
    model = Pipeline(steps=[
        ("pre", preprocessor),
        ("alg", algorithm)
    ])

    train_sizes, train_scores, test_scores = learning_curve(
        estimator=model,
        X=X_data,
        y=y_data,
        cv=3,
        scoring="r2",
        n_jobs=1,
        random_state=42
    )

    mean_train_scores = np.mean(train_scores, axis=1)
    std_train_scores = np.std(train_scores, axis=1)
    train_label = f"{mean_train_scores[-1]:.2f} +/- {std_train_scores[-1]:.2f}"

    mean_test_scores = np.mean(test_scores, axis=1)
    std_test_scores = np.std(test_scores, axis=1)
    test_label = f"{mean_test_scores[-1]:.2f} +/- {std_test_scores[-1]:.2f}"

    fig, ax = plt.subplots(figsize=figsize)

    # Training curve
    plot_curve(
        train_sizes,
        mean_train_scores,
        std_train_scores,
        f"Train ({train_label})",
        ax
    )

    # Testing curve
    plot_curve(
        train_sizes,
        mean_test_scores,
        std_test_scores,
        f"Test ({test_label})",
        ax
    )

    ax.set(xlabel="Training set size", ylabel="R-squared", title=name)
    ax.legend(loc ="lower right")
    plt.show()

```

```python
for name, alg in algorithms.items():
    plot_learning_curve(name, alg)
```

##model traininng

```python
model = Pipeline(steps= [
    ("pre", preprocessor),
    ("rf",RandomForestRegressor(n_estimators = 10))
])
```

```python
model.fit(X_data,y_data)
```

##model evaluation

```python
def evaluate_model (X,y):
    y_pred = model.predict(X)
    return r2_score(y,y_pred)
    
```

```python
print(f"R2 score on training data is =({evaluate_model (X_data,y_data)}")
```

```python
print(f"R2 score on training data is =({evaluate_model (X_test,y_test)}")
```
